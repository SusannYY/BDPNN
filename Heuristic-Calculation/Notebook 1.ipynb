### Import Earthquake and calculate risks

```
earthquake_risk = pd.read_csv('earthquake_censustract.csv')
earthquake_risk
```

```
# assign weights to the variables
w_distance = 0.3
w_intensity = 0.5
w_magnitude = 0.2

# calculate the distance from populated areas (using a buffer of 10km)
buffer = 0.1 # 10km in degrees
earthquake_risk['distance'] = np.sqrt((earthquake_risk['Longitude'] - earthquake_risk['Longitude'].mean())**2 + (earthquake_risk['Latitude'] - earthquake_risk['Latitude'].mean())**2)
earthquake_risk['distance'] = np.where(earthquake_risk['distance'] > buffer, buffer, earthquake_risk['distance'])

# calculate the expected MMI at a given location
'''
MMI stands for "Modified Mercalli Intensity". 
It is a measure of earthquake intensity that is based on the effects 
of the earthquake on people, buildings, and the environment. 
'''
def calculate_mmi(row):
    return 1.5 * row['Magnitude'] - 3.3 * np.log10(row['Depth']) - 2.92 * np.log10(row['distance']) + 8.0
earthquake_risk['mmi'] = earthquake_risk.apply(calculate_mmi, axis=1)

# calculate the earthquake risk score
earthquake_risk['risk_score'] = w_distance * earthquake_risk['distance'] + w_intensity * earthquake_risk['mmi'] + w_magnitude * earthquake_risk['Magnitude']
```

```
max_earth_risk = max(earthquake_risk['risk_score'])
min_earth_risk = min(earthquake_risk['risk_score'])
earthquake_risk['Normalized Earthquake Risk'] = ((earthquake_risk['risk_score'] - min_earth_risk)/(max_earth_risk - min_earth_risk)).round(2)
earthquake_risk
```

```
earthquake_risk_by_census_tract = earthquake_risk.groupby('spa_names')['Normalized Earthquake Risk'].mean()
earthquake_risk_by_census_tract_df = pd.DataFrame(earthquake_risk_by_census_tract)
```

```
earthquake_risk_by_census_tract_df['Normalized Earthquake Risk'] = (1 - earthquake_risk_by_census_tract_df['Normalized Earthquake Risk']).round(2)
```

### Census Tracts to improve outreach

- Member Normal: normalized amount of members from original contact database (45%)

- Experience Percentage: average of experience per census tract (20%)

- Yes Percentage: average percentage of participants willing have future participation (25%)

- Normalized Fire Risk: inverted, where lower value means higher risk (10%)

```
by_census_tract_combined
```

```
heuristic_df = by_census_tract_combined
```

```
# Heuristic calculation
heuristic_df['heuristic'] = (heuristic_df['Member Normal'] * 0.45 
    + heuristic_df['Experience Percentage'] * 0.2 
    + heuristic_df['Yes Percentage'] * 0.25 
    + heuristic_df['Normalized Fire Risk'] * 0.05 
    + heuristic_df['Normalized Earthquake Risk']*0.05).round(2)
```

Order of census tracts to implement outreach

```
max_val = max(heuristic_df['heuristic'])
min_val = min(heuristic_df['heuristic'])

heuristic_df['normalized'] = ((heuristic_df['heuristic'] - min_val) / (max_val - min_val)).round(2)
heuristic_df[['normalized']]
```

We've come up with a list of census tracts that can be used for determining the areas BDPNN should improve its outreach efforts.
